```python
#Drop unnecessary columns
df_train.drop(["Product_id","Customer_name"],axis=1,inplace=True)
df_test.drop(["Product_id","Customer_name"],axis=1,inplace=True)
```

```python
#Convert instock date to datetime
df_train["instock_date"]=pd.to_datetime(df_train["instock_date"])
df_test["instock_date"]=pd.to_datetime(df_test["instock_date"])
```

```python
for df in [df_train,df_test]:
    df["Year"]=df["instock_date"].dt.year
    df["Month"]=df["instock_date"].dt.month
    df["Day"]=df["instock_date"].dt.day
    df["Hour"]=df["instock_date"].dt.hour
```

```python
df_train.drop("instock_date",axis=1,inplace=True)
df_test.drop("instock_date",axis=1,inplace=True)
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
#Encode categorical values for ordinal data
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df_train["Loyalty_customer"]=LabelEncoder().fit_transform(df_train["Loyalty_customer"])
df_test["Loyalty_customer"]=LabelEncoder().fit_transform(df_test["Loyalty_customer"])
```

```python
#Encode categorical values for nominal data
df_train=pd.get_dummies(df_train,columns=["Product_Category"],drop_first=True)
df_test=pd.get_dummies(df_test,columns=["Product_Category"],drop_first=True)
```

```python
df_train.info()
```

```python
#Convert boolean values into int
bool_cols=df_train.select_dtypes(include=["bool"]).columns
bool_cols=df_test.select_dtypes(include=["bool"]).columns
df_train[bool_cols]=df_train[bool_cols].astype(int)
df_test[bool_cols]=df_test[bool_cols].astype(int)
```

```python
#handle missing values in numerical data
num_col = df_train.select_dtypes(include=["float64","int64"]).columns
df_train[num_col]=df_train[num_col].fillna(df_train[num_col].median())

num_col = df_test.select_dtypes(include=["float64","int64"]).columns
df_test[num_col]=df_test[num_col].fillna(df_test[num_col].median())
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
#Feature Engineering 
df_train["Price_Range"]=df_train["Maximum_price"]-df_train["Minimum_price"]
df_test["Price_Range"]=df_test["Maximum_price"]-df_test["Minimum_price"]
```

```python
#Separate Features and target
x=df_train.drop("Selling_Price",axis=1)
y=df_train["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split
#Train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#Train machine learning model
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
#Model Evaluation
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
#Predict on test data
y_preds=model.predict(x_test)
```

```python
#Calculate evaluation metrics
mae=mean_absolute_error(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_preds)
```

```python
print(" Mean Absolute Error :", mae)
print(" Mean Squared Error :", mse)
print(" Root Mean Squared Error :", rmse)
print(" R² Score:", r2)
```
